```python
from ultralytics import YOLO
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
model = YOLO('yolov8s-cls.pt')
```

```python
model.train(data='./data/YOLOv8_dataset', epochs=25, workers=8, name='train', exist_ok=True)
```

```python
%matplotlib inline
results_csv = pd.read_csv('./runs/classify/train/results.csv')
plt.plot(results_csv['                  epoch'], results_csv['             train/loss'], results_csv['               val/loss'])
plt.legend(['Training loss', 'Validation loss'])
plt.title('Train and Validation losses over 25 epochs')
plt.savefig('./runs/classify/train/train-val_loss.jpg', dpi=150)
plt.show()
```

```python
test = pd.read_csv('./data/YOLOv8_dataset/test.csv')
```

```python
preds = model.predict('./data/YOLOv8_dataset/Test')
```

```python
labels = preds[0].names
```

```python
y_pred = []
for pred in preds:
	y_pred.append(labels[pred.probs.top1])

y_pred[:10]
```

```python
y_true = test['minifigure_name'].values.tolist()
```

```python
accuracy_score(y_true, y_pred)
```
